```python
import pandas as pd
```

```python
df = pd.read_csv('../data/Tesla_news.csv')
df.head()
```

```python
df.info()
```

```python
# drop duplicated ones
df = df.drop_duplicates()
df.info()
```

```python
# randomly extract 60% of data for mannual labeling
sample_size = int(df.shape[0] * 0.6)
df_select = df.sample(n = sample_size, random_state = 1)
df_select.head()
```

```python
df_select.info()
```

```python
df_unselect = df.drop(df_select.index)
df_unselect.info()
```

```python
df_unselect.head()
```

```python
df_unselect.reset_index(drop = True, inplace = True)
df_unselect.head()
```

```python
df_unselect.tail()
```

```python
df_select.sort_values(by = ['date'], inplace = True)
```

```python
df_select.reset_index(drop = True, inplace = True)
df_select.head()
```

```python
df_select.tail()
```

```python
# saveout the data
df_select.to_csv('../data/Tesla_news_with_labels.csv', index = False)
df_unselect.to_csv('../data/Tesla_news_without_labels.csv', index = False)
```
